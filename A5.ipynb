```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Question 2

# A)

```python
T = 20    # seconds
Fs = 1000 # Hz
t = np.arange(T*Fs) / Fs # time in seconds
x = np.zeros(T*Fs)
for i in range(len(x)):
    if(np.random.rand() < 0.0005):
        x[i] = 1
plt.plot(t, x)
plt.title('Simulated Spike Train')
plt.xlabel('Time (sec)')
plt.ylabel('Is Spiking')
```

# B)

```python
t_rise = 0.005  # seconds
t_decay = 0.150 # seconds
h = (1 - np.exp(-t/t_rise))*np.exp(-t/t_decay)
# Using convolution theorem
CaTrace = np.real(np.fft.ifft(np.multiply(np.fft.fft(x), np.fft.fft(h))))
plt.plot(t, CaTrace)
```

# C)

```python
noiseRMS = 0.01 # arbitrary units (uV?)
CaTrace += np.random.normal(0, noiseRMS, len(t))
plt.plot(t, CaTrace)
```

# D)

```python
H = np.fft.fft(h)
magH = 20*np.log10(np.real(H*np.conj(H)))
plt.xlim(0,50)
plt.plot(magH)
```
