```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# Problem 1

```python
t0 = np.pi
k = 1
I = 1 # should be some function of k
t = np.arange(-3,20,0.01)
P = np.exp(k*np.cos(t-t0))*I/(2*np.pi)

#
a0 = 2*np.mean(P)

# plot the von Mises mass function and its mean
plt.plot(t,P,t,a0/2*np.ones(len(t)))
```
