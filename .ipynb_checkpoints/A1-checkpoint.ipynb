```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Plot original simulated data (30 Hz sinusoid)
# The Nyquist limit is 30*2 or 60 Hz
Fs = 240  # Hz
T = 1     # seconds
t = np.arange(0,1,1/Fs*T)
simdata = np.sin(2*np.pi*30*t)
plt.plot(t, simdata)
```

```python
# Downsample 2x (120 Hz Sampling rate)
# The data looks the same as before (maybe it's a tiny bit sharper, but it's hard to tell).
t2 = t[::2]
simdata2 = np.sin(2*np.pi*30*t2)
plt.plot(t2, simdata2)
```

```python
# Downsample 3x (80 Hz Sampling rate, nearing the Nyquist limit)
# The data already looks strange. It still appears periodic around 30 Hz, but 
# with a decreased amplitude on 2 out of 3 peaks.
t3 = t[::3]
simdata3 = np.sin(2*np.pi*30*t3)
plt.plot(t3, simdata3)
```

```python
# Downsample 4x  (60 Hz Sampling rate, at the Nyquist limit)
# This graph looks like noise.  It is likely rounding error mixed with the artifact
# of sampling when the wave is very close to zero.
t4 = t[::4]
simdata4 = np.sin(2*np.pi*30*t4)
plt.plot(t4, simdata4)
```

```python
# Downsample 8x (30 Hz sampling rate, well below the Nyquist limit)
# We are now at the frequency of the source itself.  Again, the noisy signal is
# likely due to rounding error and sampling when the source is close to zero.
t8 = t[::8]
simdata8 = np.sin(2*np.pi*30*t8)
plt.plot(t8, simdata8)
```

# 4d visualization

```python
x = np.arange(-10, 10, 0.01)
S = np.zeros(len(x))
for n in range(1, 5):
    S = S + np.sin(n*x)/n
    plt.plot(x, S)
```
