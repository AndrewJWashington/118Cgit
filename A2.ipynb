# Problem 1

```python
t = np.arange(-6.28,6.28,0.01)
t0 = np.pi   # phase shift constant
k = 1        # width constant
I = np.i0(k) # normalization constant
P = np.exp(k*np.cos(t-t0))*I/(2*np.pi) # von Mises Distribution

#
a0 = 2*np.mean(P)
print(I)
print("The numerical value for a0 is: " + str(a0))
print("The analytic value for a0 is:  " + str(1/np.pi))

# plot the von Mises mass function and its mean
plt.plot(t,P,t,(a0/2)*np.ones(len(t)))
```
